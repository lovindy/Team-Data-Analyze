```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
pwd = os.getcwd()
df_ori = pd.read_csv(pwd + "/US Dollar Index Historical Data.csv");
df_ori.head()
```

```python
df = df_ori.copy()
```

```python
df['Date'] = pd.to_datetime(df['Date'] ,format='%b %d, %Y')
```

```python
df.head(10)
```

```python
df['Change %'] = df['Change %'].str.replace('%','').astype(float)
df.head(10)
```

```python
season_ = df.set_index('Date')
season_.head()
```

```python
plt.figure(figsize=(15,10))
plt.plot(season_['Price'],linewidth=1,color='orange')
plt.show()
```

```python
season_.head(10)
```

```python
season_.info()
```

```python

season_['Month'] = season_.index.month
season_['Year'] = season_.index.year
season_.head()
```

```python
import plotly.graph_objects as go
from datetime import datetime

layout = go.Layout(
   title = 'Price',
   yaxis = dict(
       autorange = True,
       fixedrange= False))
    
fig = go.Figure(data=[go.Candlestick(x=season_.index,
                open=season_['Open'],
                high=season_['High'],
                low=season_['Low'],
                close=season_['Price'])],
                layout=layout)


fig.show()
```

```python



# Python3 code to demonstrate 
# Getting a tuple of ISO Year,  
# ISO Week Number and ISO Weekday 
  
# Importing date module from datetime 
from datetime import date 
   
# Calling the today() function 
# to return todays date 
Todays_date = date.today() 
  
# Printing today's date 
print(Todays_date) 
  
# Calling the isocalendar() function 
# over the above today's date to return 
# its ISO Year, ISO Week Number 
# and ISO Weekday 
print(Todays_date.isocalendar())
```
