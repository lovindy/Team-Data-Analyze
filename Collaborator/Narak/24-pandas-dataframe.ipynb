<h2>Create a DataFrame</h2>

```python
import pandas as pd

data = {'col_1' : [3, 2, 1, 0], 'col_2' : ['a', 'b', 'c', 'd'] }
df = pd.DataFrame(data)
df
```

```python
s1 = pd.Series([1.0, 2.0, 3.0], index = ['a' , 'b', 'c'])
s2 = pd.Series([1.0, 2.0, 3.0, 4.0], index = ['a', 'b', 'c', 'd'])

data = {'one' : s1, 'two' : s2}

pd.DataFrame(data)
```

```python
# Create a dataframe
d = {'col1' : [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data = d)
print(df)

# Rename the index
df.index = ['new index0', 'new index1']
print(df)
# Rename the column
df.columns = ['new_col1', 'new_col2']
print(df)
```

```python
d = {'col1' : [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data = d)
df.rename(index = {0 : 'New index0'}, inplace=True)
df.rename(columns = {'col2': 'new_col2'})
```

```python
animals = {'Animal' : ['Alligator', 'Bee', 'Falcon', 'Lion', 'Monkey']}
df = pd.DataFrame(animals, index=range(1, len(animals['Animal']) + 1))

df.head() # 5 rows
df.shape # (5, 1)
len(df) # 5 rows
df.columns # ['Animal']
df.index # RangeIndex(start=1, stop=6, step=1)
df.values # [[ 'Alligator'], ['Bee'], ['Falcon'], ['Lion'], ['Monkey']]
```
