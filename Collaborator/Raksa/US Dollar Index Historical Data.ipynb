```python
import plotly.express as px
def plotting (title, data, x, y, x_label, y_label):
    # generating function to plot the data
    # data: DataFrame containing the data to plot.
    # x: Column name to use for the x-axis.
    # y: Column name to use for the y-axis.
    # x_label: Label for the x-axis.
    # y_label: Label for the y-axis.
    fig = px.line(data, x= x, y=y, labels={x: x_label, y: y_label})
    
    fig.update_layout(template='simple_white', font=dict(size=18),
                      title_text=title, width=650, title_x=0.5, height=400)
    fig.show()
```

```python
plotting(title='US Dollar Index Historical Data',
         data=df_md,
         x='Price',
         y='Date',  # Make sure 'Value' matches your DataFrame's column
         x_label='US Dollar Index Price',
         y_label='Date')
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(30, 15))
plt.plot(df_md['Date'], df_md['Price'], label='US Dollar Index')
plt.title('US Dollar Index Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
