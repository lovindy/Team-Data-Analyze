```python
# remove the column that we don't need

df = df.drop(['Unnamed: 0', 'Uri', 'Liveness', 'Valence', 'Channel', 'Instrumentalness', 'Acousticness', 'Url_spotify', 'Speechiness', 'Loudness', 'Duration_ms', 'Url_youtube', 'Description', 'Licensed', 'official_video', 'Stream'], axis=1)
```

```python
df
```

```python
# find the artist with the most liked songs

df.groupby('Artist')['Likes'].sum().sort_values(ascending=False)

df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
# keep only 100 rows
df = df.head(50)
```

```python
df
```

```python
# find the most like artist
# the groupby is used to find the artist with the most liked songs, and the sum is used to find the total number of likes
# the sort_values is used to sort the dataframe in descending order

df.groupby('Artist')['Likes'].sum()
```

```python
# find 5 of the most like track 

the_most_likes_track = df.groupby('Track')['Likes'].sum().sort_values(ascending=False).head(20)

the_most_likes_track
```

```python
the_most_likes_track.to_excel('the_most_like_track.xlsx')
```
