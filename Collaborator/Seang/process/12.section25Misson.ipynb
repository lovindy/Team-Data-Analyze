```python
# step 6 coefficient of variation (CV)
# Coefficient of variation = Standard deviation / Mean
study_time_cv = df['studytime'].std() / df['studytime'].mean()

print(study_time_cv)
```

```python
study_time_cv = df['absences'].std() / df['absences'].mean()

print(study_time_cv)
```
