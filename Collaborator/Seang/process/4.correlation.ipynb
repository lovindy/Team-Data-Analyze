```python
import pandas as pd

# Create a DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [1, 1, 2, 2, 3]
}
df = pd.DataFrame(data)

# Compute correlation matrix
corr_matrix = df.corr()
print(corr_matrix)
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(5, 60, 10)
y = [38, 22, 32, 14, 20, 5]
```

```python
fig, ax = plt.subplots()

ax.scatter(x, y)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot')

plt.show()
np.corrcoef(x, y)
```
