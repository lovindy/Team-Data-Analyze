```python
# 1. convert dictionary data into series object
# 2. Calculate the population fluctations for each city and represent oit as a bar graph
# 3. Calculate the population density base on the capital city of each country and represent it using visualization tool
# 4. Find the citiest with the highest and lowest population density

# import library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# population in 2020
population_2020 = {
    'Tokyo': 3773892,
    'New York': 8175133,    
    'Beijing': 21513926,
    'Seoul': 9731488,
    'Manila': 1009900,
    'Phnom Penh': 2301372,
    'Moscow': 11979529, 
    'Shanghai': 24256800,
    'Istanbul': 1548195,
    'Delhi': 28098000,
    'Karachi': 14542313,
    'Mumbai': 22140000,
    'Shenzhen': 1477309,
    'Saigon': 11776000,
    'Bangkok': 21670000,
    'Lagos': 13420000,
    'Jakata': 9500000,
}

# population in 2021
population_2021 = {
    'Tokyo': 37738212,
    'New York': 81752133,    
    'Beijing': 215133926,
    'Seoul': 97314188,
    'Manila': 10099300,
    'Phnom Penh': 23011372,
    'Moscow': 119795229, 
    'Shanghai': 242526800,
    'Istanbul': 15438195,
    'Delhi': 280980600,
    'Karachi': 145542313,
    'Mumbai': 221406000,
    'Shenzhen': 14773029,
    'Saigon': 117760600,
    'Bangkok': 216708000,
    'Lagos': 134200040,
    'Jakata': 95008000,
}

# city_area
city_area = {
   'Tokyo': 2194,
   'New York': 3025,
   'Beijing': 1684,
   'Seoul': 6051,
   'Manila': 2395,
   'Phnom Penh': 2395,
   'Moscow': 3425,
   'Shanghai': 2255,
   'Istanbul': 2155,
   'Delhi': 2155,
   'Karachi': 2155,
   'Mumbai': 2155,
   'Shenzhen': 2155,
   'Saigon': 2155,
   'Bangkok': 2155,
   'Lagos': 2155,
   'Jakata': 2155,
}

# convert the diction into series object
population_2020 = pd.Series(population_2020)
population_2021 = pd.Series(population_2021)
city_area = pd.Series(city_area)

population_density_2021 = population_2021 / city_area
population_density_2020 = population_2020 / city_area

# find the growth
population_growth = population_density_2021 - population_density_2020

print(population_growth.sort_values())
```

```python
plt.figure(figsize=(13, 6))
plt.xticks(rotation='vertical', size = 10)
plt.bar(population_growth.index, population_growth)
plt.show()
```

```python
plt.figure(figsize=(13, 6))
plt.xticks(rotation=90, size = 10)
plt.barh(population_growth.sort_values().index, population_growth.sort_values())
plt.show()
```

```python
plt.figure(figsize=(13, 6))
plt.xticks(rotation='vertical', size = 10)
plt.bar(population_growth.sort_values().index, population_growth.sort_values(), color = 'pink')
plt.show()
```
