```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
tips = sns.load_dataset("tips")

# Compute the correlation matrix
numeric_columns = tips.select_dtypes(include=['float64', 'int64']).columns

# Create a new DataFrame with only numeric columns
numeric_tips = tips[numeric_columns]

# Calculate correlation matrix
corr = numeric_tips.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Heatmap of Correlations in Tips Dataset')
plt.show()
```

```python
import numpy as np
import pandas as pd

# Generate sample data
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)
data = pd.DataFrame({'x': x, 'y': y})

# Compute the correlation matrix
corr = data.corr()

# Plot the heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Strong Positive Correlation')
plt.show()
```

```python
y = -2 * x + np.random.normal(0, 0.1, 100)
data = pd.DataFrame({'x': x, 'y': y})

# Compute the correlation matrix
corr = data.corr()

# Plot the heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Strong Negative Correlation')
plt.show()
```
