```python
# the ISO-8859-1 encoding is used for the latin-1 encoding
# because the latin-1 encoding is not supported in Windows
spotify_2019 = pd.read_csv('../data/top50.csv', encoding = 'ISO-8859-1')
spotify_2019
```

```python
# 1 Based on this dataset, find the artist with most popular songs
# 2 Verify if there is a correlation between tempo of the music and its popularity through data visualization
# 3 Finally, create a tracklist consisting only of track from your favorite artist

# Start the process
spotify_2019.head(10)
```

```python
spotify_2019.info()
```

```python
spotify_2019.describe()
```

```python
# rename the Unname to rank

spotify_2019 = spotify_2019.rename(columns = {'Unnamed: 0': 'Rank'})
```

```python
# select only the necessary columns to create a new DataFrame

# Unnamed: 0	Track.Name	Artist.Name	Genre	Beats.Per.Minute	Energy	Danceability	Loudness..dB..	Liveness	Valence.	Length.	Acousticness..	Speechiness.	Popularity
df = spotify_2019[['Rank', 'Track.Name', 'Artist.Name', 'Beats.Per.Minute', 'Energy', 'Danceability', 'Popularity']]
df.head(10)
```

```python
# find the artist with most popular songs
popular_song = df[df["Popularity"] > 90]
popular_song.head()
```

```python
# the value is used to rank the most popular songs in descending order
# it returns as a series object
rank = popular_song['Artist.Name'].value_counts()
rank
```

```python
play_list = popular_song[popular_song["Artist.Name"] == "Post Malone"]
Malone = play_list.reset_index(drop=True, inplace=False)
Malone
```

```python
x = popular_song["Energy"]
y = popular_song["Popularity"]

plt.scatter(x, y, color="red", alpha=0.5)
plt.show()
```

```python
# correlation
# pearson is the default method, it is used when no method is specified, and it is used for numeric data
# it returns the correlation matrix
# it returns as a dataframe
# the correlation coefficient is between -1 and 1
# 1 is the highest correlation
# -1 is the lowest correlation
# numeric_only is set to True, it only returns the correlation matrix for numeric data
cor = popular_song.corr(method='pearson', numeric_only=True)
cor
```

```python
plt.matshow(cor)
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(cor, annot=True)
plt.show()
```
