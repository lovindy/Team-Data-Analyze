```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, c='blue', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2)
fig.suptitle('Multiple Subplots Example')

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Plot 1')

axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Scatter Plot 2')

axs[1, 0].bar(categories, values)
axs[1, 0].set_title('Bar Plot 3')

axs[1, 1].hist(y, bins=5)
axs[1, 1].set_title('Histogram Plot 4')

for ax in axs.flat:
    ax.set(xlabel='X-label', ylabel='Y-label')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns  # Importing the Seaborn library for statistical data visualization
import pandas as pd  # Importing the Pandas library for data manipulation

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})  # Creating a DataFrame from x and y arrays

# Create a line plot
sns.lineplot(x='x', y='y', data=data)  # Creating a line plot using Seaborn
plt.title('Line Plot')  # Adding a title to the plot
plt.show()  # Displaying the plot
```

```python

import matplotlib.pyplot as plt  # Importing the Matplotlib library for plotting

# Sample data
categories = ['A', 'B', 'C']  # List of category labels
values = [10, 15, 7]  # List of values corresponding to the categories

# Create a bar plot
plt.figure(figsize=(8, 6))  # Creating a figure with a specific size
plt.bar(categories, values, color=['red', 'green', 'blue'])  # Creating a bar plot with colors
plt.xlabel('Categories')  # Adding a label to the x-axis
plt.ylabel('Values')  # Adding a label to the y-axis
plt.title('Bar Plot')  # Adding a title to the plot
plt.show()  # Displaying the plot

```

```python
import seaborn as sns  # Importing the Seaborn library for statistical data visualization
import pandas as pd  # Importing the Pandas library for data manipulation

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})  # Creating a DataFrame from categories and values

# Create a bar plot
sns.barplot(x='categories', y='values', data=data, palette='viridis')  # Creating a bar plot using Seaborn with a color palette
plt.title('Bar Plot')  # Adding a title to the plot
plt.show()  # Displaying the plot
```

```python

import matplotlib.pyplot as plt  # Importing the Matplotlib library for plotting
import numpy as np  # Importing the NumPy library for numerical operations

# Sample data
data = np.random.randn(1000)  # Creating an array of 1000 random values from a standard normal distribution

# Create a histogram
plt.figure(figsize=(8, 6))  # Creating a figure with a specific size
plt.hist(data, bins=30, color='purple', alpha=0.7)  # Creating a histogram with 30 bins and color
plt.xlabel('Value')  # Adding a label to the x-axis
plt.ylabel('Frequency')  # Adding a label to the y-axis
plt.title('Histogram')  # Adding a title to the plot
plt.show()  # Displaying the plot

```

```python

import seaborn as sns  # Importing the Seaborn library for statistical data visualization

# Create a histogram
sns.histplot(data, bins=30, color='purple')  # Creating a histogram using Seaborn with 30 bins and color
plt.title('Histogram')  # Adding a title to the plot
plt.show()  # Displaying the plot

```
