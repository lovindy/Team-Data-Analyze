# 3) Data Cleaning

```python
df = df_ori.copy()
```

```python
df.head()
```

```python
# find mode of column key and fill the mode of column key top the NaN data
mode_key = df['key'].mode()[0]

df['key'].fillna(mode_key, inplace = True)
df['key'].isnull().sum()
```

```python
# sort data of streams
df['streams'].sort_values(ascending = False)
# unable to sort yet, cuz the data isn't "int64" but object
```

We see that index of 574 has a bug data (not int like other)

```python
# remove that row
df.drop(574, inplace = True)
```

```python
df['streams'].sort_values(ascending = False)
```

```python
# convert data type
df['streams'] = df['streams'].astype('int64')
```

```python
df['streams'].dtype
```

```python
# sort the value of streams from big to small
df.sort_values(by = 'streams', ascending = False, inplace = True)
df['streams']
```

```python
df.head()
```

```python
# reset index of data frame

df.reset_index(drop = True, inplace = True)
df
```
