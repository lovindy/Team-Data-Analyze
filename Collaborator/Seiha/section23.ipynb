```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

# Viewing Data
print(df.head())

# Adding a column
df['Salary'] = [70000, 80000, 90000]

# Filtering Data
adults = df[df['Age'] > 30]
print(adults)

# Saving to CSV
df.to_csv('data_with_salary.csv', index=False)

# Loading from CSV
df_new = pd.read_csv('data_with_salary.csv')
print(df_new)
```
