"DataFrame".dropna(axis = 1, how = 'any', thresh = None, subset = None, inplace = False)
+ Delet columns.
    - we previously confirmed that out of the total 891 passengers, 688 of them have missing 'deck' info.
    - considering the high proportion of missing daat, it may be dem,anded irrelevant for data processing and analysis. in this case, the most common method of handling missing data is to delete column with missing data.

```python
new_df = df.dropna(axis = 1, thresh = 500)

new_df.head()

# thresh = 500 : delete columns that have 500 or more NaN values.
# since 'deck' column has 688 Nan values, which is more than 500, now it's completedly deleted
```

"DataFrame".dropna(axis = 0, how = 'any', thresh = None, subset = None, inplace = False)
- Delete rows.
    - out of the 981 pass, 177 of them have missing age data. If the age of the pass is considered an important
      variable for data analysis, it is recommended to removed the row (data) of passengers with missing age data.
    - By using subset = 'age', all rows (axis=0) in the age column with NaN values are deleted.
    - how='all' only delete rows if all the data of a row is NaN.

```python
age_df = df.dropna(axis = 0, how='any', subset = ['age']) # deleting the rows with missing data in age column
len(age_df) # see now the length is shorter
```
