```python
import pandas as pd

df = pd.read_csv("covid-vaccination-doses-per-capita.csv")
df.head()
```

```python
df.info()
```

```python
# create a new column "Date" inhert data from "Day"
df['Date'] = pd.to_datetime(df['Day'])

# set index of df using the Date column
df.set_index("Date", inplace = True)

# drop the column "date", axis = 1(vertical)
df.drop(['Day'], axis = 1, inplace = True)

df.head()
```

```python
# there are 236 countries, (unique)
len(df['Entity'].unique())
```

```python
# generate a new data frame from "Entity" column
covid_c = df.groupby('Entity')
```

```python
#distinguish the information of the created groups based on the key.
for key, group in covid_c :
    print('+key :', key) # keys of the group
    print('+number :', len(group)) # number of data points for each country grouped
    print(group.head()) # print 5 rows of each group
    print('\n')
```

```python
total_df = covid_c.sum(numeric_only = True)
total_df.head()
```

```python
import pandas as pd
# https://raw.githubusercontent.com/python-visualization/folium/master/CHANGES.txt
url = ("https://github.com/python-visualization/folium/tree/main/examples/data")
state_geo = f"{url}/us-states.json"
state_unemployment = f"{url}/US_Umemployment_Oct2012.csv"
state_data = pd.read_csv(state_unemployment)

m = folium.Map(location = [48, -102], zoom_start = 3)

folium.Choropleth(
    geo_data = state_geo,
    name = "choropleth",
    data = state_data,
    columns = ["State", "Umemployement"],
    key_on = "feature.id",
    fill_color = "YlGn",
    fill_opacity = 0.7,
    line_opacity = 0.2,
    legend_name = "Umemployment Rate (%)",
).add_to(m)

folium.LayerControl().add_to(m)
m
```

```python
import folium
import json

center = [11.59562933746947, 104.92580536174557]

m = folium.Map(location = center, zoom_start = 2,
               max_bounds = True,
               min_zoom = 1, min_lat = -84,
               max_lat = 84, min_lon = -175, max_lon = 187,
               )

geo_path = "./data/World_Countries_Generalized.geojson"

json_data = json.load(open(geo_path, encoding = 'utf-8'))

folium.Choropleth(geo_data = json_data,
                  data = total_df,
                  columns = (total_df.index, 'total_vaccinations_per_hundred'),
                  key_on = 'properties.COUNTRY',
                  fill_color = 0.7,
                  line_opacity = 0.7,
                  fill_opacity = 0.5,
).add_to(m)

folium.LayerControl().add_to(m)
```
